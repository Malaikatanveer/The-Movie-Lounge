```python
import pandas as pd
import spacy
import re

#spaCy English model
nlp = spacy.load("en_core_web_sm")

df = pd.read_csv("metacritic-reviews.csv", encoding='windows-1252')

if 'Unnamed: 0' in df.columns:
    df = df.iloc[1:]

df.dropna(subset=['summary'], inplace=True)


#preprocess text
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    return " ".join(tokens)

preproceesed_records= df['summary'].apply(preprocess_text)
print(preproceesed_records)
```

```python
from collections import defaultdict

inverted_index = defaultdict(lambda: defaultdict(list))

#inverted index
def create_inverted_index(records):
    for doc_idx, record in enumerate(records):
        words = record.split()
        for word_idx, word in enumerate(words):
            inverted_index[word][doc_idx].append(word_idx)
    # positions in each document
    for word, doc_positions in inverted_index.items():
        for doc_idx, positions in doc_positions.items():
            inverted_index[word][doc_idx] = sorted(positions)
    return inverted_index

inverted_index = create_inverted_index(preproceesed_records)
for word, doc_positions in inverted_index.items():
    print(f"{word}: {doc_positions}")
```

```python
import tensorflow as tf
import tensorflow_hub as hub

# Load ELMo model
elmo = hub.load("https://tfhub.dev/google/elmo/3")

def get_elmo_embedding(text):
    embeddings = elmo.signatures["default"](tf.constant(text))["elmo"]
    return embeddings.numpy()[0]

# ELMo embeddings for each word in the corpus
def get_elmo_embeddings(preprocessed_records):
    elmo_embeddings = {}
    for record in preprocessed_records:
        for word in record.split():
            if word not in elmo_embeddings:
                elmo_embeddings[word] = get_elmo_embedding([word])
    return elmo_embeddings

elmo_embeddings = get_elmo_embeddings(preproceesed_records)
sample_word = "example"
print(f"Embeddings for '{sample_word}': {elmo_embeddings[sample_word]}")
```

```python
import requests
from flask import Flask, jsonify, request
from pyngrok import ngrok
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from flask_cors import CORS


app = Flask(__name__)
CORS(app)

# Function to preprocess text
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    return " ".join(tokens)
    return text

@app.route('/api/movies')
def get_movies():
    
#     print(movie_names)
#     print(release_dates)
#     print(ratings)
    
    movies = []
    
    for i in range(len(movie_names)):
        movie = {
            'name': movie_names[i],
            'release_date': release_dates[i],
            'rating': ratings[i],
            'summary': summaries[i],
            'user_rating': user_ratings[i],
            'website_rating': website_ratings[i]
        }
        movies.append(movie)
        
    print(len(movies))
        
    return jsonify(movies)

# Endpoint to handle search query
@app.route('/api/search', methods=['POST'])
def search_movie():
    global movie_names, release_dates, ratings, summaries, user_ratings, website_ratings
    movie_names = []
    release_dates = []
    ratings = []
    summaries = []
    user_ratings = []
    website_ratings = []
    
    search_data = request.get_json()
    search_query = search_data.get('query')  
    print(search_query)

    # Preprocess the search query
    processed_query = preprocess_text(search_query)
    threshold = 0.7
    similar_words_for_phrase = {}

    for word in processed_query.split():
        # Retrieve embeddings
        other_words = [w for w in processed_query.split() if w != word]
        other_embeddings = [elmo_embeddings[w] for w in other_words]

        # Calculate average embedding 
        avg_other_embedding = np.mean(other_embeddings, axis=0)

        # Find similar words for the current word
        similar_words_for_phrase[word] = []
        for other_word in processed_query.split() + list(elmo_embeddings.keys()):
            if other_word != word and other_word not in similar_words_for_phrase[word]:
                other_word_embedding = elmo_embeddings[other_word]
                similarity = cosine_similarity(avg_other_embedding.reshape(1, -1), other_word_embedding.reshape(1, -1))[0][0]
                if similarity >= threshold:
                    similar_words_for_phrase[word].append(other_word)

    # Get all documents from the inverted index
    all_documents = list(inverted_index.values())

    # Rank the documents based on the presence of similar words
    document_scores = {}
    for word in processed_query.split():
        for doc_id, positions in inverted_index[word].items():
            if doc_id not in document_scores:
                document_scores[doc_id] = 0
            document_scores[doc_id] += len(positions) * len(similar_words_for_phrase[word])

    # Sort the documents by score and show the top 10
    top_documents = sorted(document_scores.items(), key=lambda x: x[1], reverse=True)[:10]
    
    for doc_id, score in top_documents:
        print(f"Document ID: {doc_id}, Score: {score}")

        # Retrieve information for the document
        movie_name = df.loc[doc_id, "Movie name"] if not pd.isnull(df.loc[doc_id, "Movie name"]) else "Not Available"
        release_date = df.loc[doc_id, "Release Date"] if not pd.isnull(df.loc[doc_id, "Release Date"]) else "Not Available"
        rating = re.sub(r'\|', '', df.loc[doc_id, "Rating"].strip()) if not pd.isnull(df.loc[doc_id, "Rating"]) else "Not Available"
        summary = df.loc[doc_id, "summary"] if not pd.isnull(df.loc[doc_id, "summary"]) else "Not Available"
        user_rating = df.loc[doc_id, "User rating"] if not pd.isnull(df.loc[doc_id, "User rating"]) else "Not Available"
        website_rating = df.loc[doc_id, "Website rating"] if not pd.isnull(df.loc[doc_id, "Website rating"]) else "Not Available"

        # Append data to respective lists
        movie_names.append(movie_name)
        release_dates.append(release_date)
        ratings.append(rating)
        summaries.append(summary)
        user_ratings.append(user_rating)
        website_ratings.append(str(website_rating))


    return jsonify({"message": "Successful"})


if __name__ == '__main__':
    app.run()

# Set up a tunnel to the Flask app on port 5000
public_url = ngrok.connect(port='8888')
print(f"Flask app is running at {public_url}")
```
